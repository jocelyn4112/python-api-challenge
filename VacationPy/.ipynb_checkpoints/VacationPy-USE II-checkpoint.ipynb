### Store Part I results into DataFrame

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api)

locations = vacation_df[["Lat","Long"]]
weights = vacation_df["Humidity"]
fig = gmaps.figure()
fig.add_layer(gmaps.heatmap_layer(locations = locations,weights = weights,max_intensity=None, dissipating=True, point_radius=None, opacity=0.6, gradient=None))
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
 #A max temperature lower than 80 degrees but higher than 70. #* Wind speed less than 10 mph #* Zero cloudiness

criteria_new_df = vacation_df.loc[(vacation_df["MaxTemp"] > 70) & (vacation_df["MaxTemp"] < 80) & (vacation_df["Cloundiness"] == 0) & (vacation_df["Wind Speed"] < 10)]
#criteria_new_df.dropna(how='any')
criteria_new_df.reset_index(inplace=True)
criteria_new_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
#new variable
hotel_df = criteria_new_df
hotel_df.head()
```

```python
#add coulumn, rename "Hotel Name"
hotel_df['Hotel Name'] = ""

hotel_df
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations = hotel_df[["Lat","Long"]]



```

```python
hotels = hotel_df["Hotel Name"]
```

```python
# geocoordinates 
target_coordinates = locations
target_search = "city"
target_radius = 5000
target_type = "hotel"

lat = target_coordinates['Lat']
long = target_coordinates['Long']
lat_lng =  zip(lat,long)

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": api
}

    
# base url
base_url = f'https://maps.googleapis.com/maps/api/place/nearbysearch/json?radius=5000&type=lodging&key={api}&location='


for index,lat_long in enumerate(lat_lng):
    
    
    url = f"{base_url}{lat_long[0]},{lat_long[1]}"
    #print(url)
    
    try:
        response=requests.get(url).json()
        hotel_name = response['results'][0]['name']
        print(hotel_name)
        hotel_df.loc[index,'Hotel Name'] = hotel_name
        
    
    except:
        print("Hotel not found")
        hotel_df.loc[index,'Hotel Name'] = np.nan
            
    # convert response to json
    #hotel_data = response.json()

    # Print the json (pretty printed)
    #print(json.dumps(hotel_data, indent=4, sort_keys=True))

```

```python
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{city}</dd>
<dt>Country</dt><dd>{city}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Long"]]
```

```python
# Create a hotel marker layer
fig = gmaps.figure()
#weights = "Humidity"

hotel_layer = gmaps.marker_layer(locations,info_box_content=hotel_info)

fig.add_layer(hotel_layer)

fig
```
